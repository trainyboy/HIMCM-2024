# Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Define AHP and Entropy Weights

```python
# Define AHP weights as a pandas Series
ahp_weights = pd.Series({
    "Safety and Fair Play": 0.507079,
    "Gender Equity": 0.154587,
    "Popularity & Accessibility": 0.148062,
    "Sustainability": 0.081227,
    "Relevance & Innovation": 0.063263,
    "Inclusivity": 0.045783
})

# Define Entropy weights as a pandas Series
entropy_weights = pd.Series({
    "Safety and Fair Play": 0.146236,
    "Gender Equity": 0.090992,
    "Sustainability": 0.110827,
    "Inclusivity": 0.311455,
    "Relevance & Innovation": 0.168308,
    "Popularity & Accessibility": 0.172181
})

# Print AHP Weights
print("AHP Weights:")
for index, value in zip(ahp_weights.index, ahp_weights.values):
    print(f"{index:<30} {value:.6f}")  # Left-align and format to 6 decimal places

# Print Entropy Weights
print("\nEntropy Weights:")
for index, value in zip(entropy_weights.index, entropy_weights.values):
    print(f"{index:<30} {value:.6f}")
```

# Combine Weights Using Multiplicative Method

```python
# Ensure both Series have the same index order
ahp_weights = ahp_weights[entropy_weights.index]  # Reorder AHP weights to match entropy index

# Calculate the combined weights using multiplicative method
multiplicative_combined_weights = ahp_weights * entropy_weights  # Element-wise multiplication
multiplicative_combined_weights /= multiplicative_combined_weights.sum()  # Normalize to sum up to 1

# Print Combined Weights
print("\nCombined Weights (Multiplicative Method):")
for index, value in zip(multiplicative_combined_weights.index, multiplicative_combined_weights.values):
    print(f"{index:<30} {value:.6f}")
```

# Visualize Combined Weights

```python
# Plot Combined Weights
plt.figure(figsize=(10, 6))
multiplicative_combined_weights.plot(kind="bar", color="lightblue")
plt.title("Combined Weights (Multiplicative Method of AHP and Entropy)")
plt.ylabel("Weight")
plt.xlabel("Criteria")
plt.xticks(rotation=45)
plt.show()
```
